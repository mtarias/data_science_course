## <b>Missing Values</b>

```python
pd.isnull(data["body"])
pd.notnull(data["body"])
pd.isnull(data["body"]).values.ravel().sum()
```

Los valores que faltan en el Dataset pueden ser por 2 razones:
* Extraccion de los datos (no se han logrado extraer correctamente o no se pueden guardar en la DB)
* Recolecci√≥n de los datos (faltan datos porque nunca se lograron obtuvieron)

## <b>Borrado de datos</b>

```python
data.dropna(0, how='all')
data.dropna(0, how='any')
```

## <b>Relleno de datos</b>

```python
data["body"].fillna(0)
```

```python
data2 = data.copy()
data2["home.dest"] = data2["home.dest"].fillna("Desconocido")
data2["body"] = data2["body"].fillna(0)
data2["age"] = data2["age"].fillna(0)
data2.head(15)
```

## <b>Reemplazo por la media</b>

```python
data3 = data.copy()
data3["age"] = data3["age"].fillna(data["age"].mean())
```

## <b>Usando ffill y bfill</b>
* ffill: Pone en los nan el 1er valor que hay delante del na 
* bfill: Pone en los nan el 1er valor que hay detras del na 

```python
data["age"].fillna(method="ffill")
```

```python
data["age"].fillna(method="bfill")
```

## <b>Variables dummy</b>
Sirven para poner a nros los atributos que son categorias (ej:sex)

```python
dummy_sex = pd.get_dummies(data["sex"], prefix="sex")
```

```python
dummy_sex.head(10)
```

```python
column_names = data.columns.values.tolist()
data4 = data.copy()
data4 = data.drop(["sex"], axis=1)
data4 = pd.concat([data4, dummy_sex],axis=1)
```

```python
def createDummies(df, var_name):
    dummy_sex = pd.get_dummies(df[var_name], prefix=var_name)
    df = data.drop([var_name], axis=1)
    df = pd.concat([df, dummy_sex],axis=1)
    return df
```
